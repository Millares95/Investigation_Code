```python
# filtering the information needed
tableToUse = pandas.read_sql_query('SELECT * FROM dbo.datdembar', conn)
```

```python
# filtering by IDs
infoToUse = tableToUse[tableToUse["idprov"] == 'IJV']
print(infoToUse.head())
```

```python
# changing to pandas data frame
infoToUse = pandas.DataFrame(infoToUse)
```

```python
# shaping the data, only the last 10 years (2008 - 2017)
infoToUse = infoToUse[infoToUse["fecha"] >= '2008-01-01']
infoToUse = infoToUse[infoToUse["fecha"] <= '2017-12-31']
print(infoToUse.head())
```

```python
# removing last column
infoToUse = infoToUse.drop(columns="dem25")
print(infoToUse.head())
```

```python
# restaring the index
infoToUse.index = numpy.arange(1, len(infoToUse) + 1, 1)
print(infoToUse.head())
```

```python
# removing February 29
print("lengthBefore =", len(infoToUse)) # checking
idx_1 = infoToUse[infoToUse["fecha"] == '2008-02-29']
idx_2 = infoToUse[infoToUse["fecha"] == '2012-02-29']
idx_3 = infoToUse[infoToUse["fecha"] == '2016-02-29']
toRemove = [idx_1.index.item(), idx_2.index.item(), idx_3.index.item()]
print(toRemove)
infoToUse = infoToUse.drop(toRemove)
print("lengthAfter =", len(infoToUse)) # checking
```

```python
# filtering data needed
infoToUseFiltered = infoToUse[infoToUse['fecha'].dt.strftime('%m-%d') == '01-01']
print(infoToUseFiltered)
```

```python
import matplotlib.pyplot as plot
# restaring the index
infoToUseFiltered.index = numpy.arange(1, len(infoToUseFiltered) + 1, 1)
print(infoToUseFiltered)
# reshape the data
dataToFit = []
for k in range(1, 11):
    print(k)
    data = infoToUseFiltered.loc[k,['dem1', 'dem2', 'dem3', 'dem4', 'dem5', 'dem6', 'dem7', 'dem8', 'dem9', 'dem10', 'dem11', 'dem12', 'dem13', 'dem14', 'dem15', 'dem16', 'dem17', 'dem18', 'dem19', 'dem20', 'dem21', 'dem22', 'dem23', 'dem24']]
    data = numpy.array(data)    
    dataToFit.append(data)
dataToFit = numpy.array(dataToFit)
dataToFit = dataToFit.flatten()
dataToFit = numpy.float64(dataToFit)
plot.plot(dataToFit)
numpy.sum(dataToFit)
numpy.mean(dataToFit)
```

```python
# importing library needed
from statsmodels.tsa.statespace.sarimax import SARIMAX

# creating a function for fitting SARIMAX
def modelSARIMAX(dataToFit):
     # creating the object model SARIMAX(0,1,1) x (0,1,1)24
    model = SARIMAX(dataToFit,order=(0,1,1),seasonal_order=(0,1,1,24))
    model_fit = model.fit() # fit model
    return model_fit
```

```python
# calling the function
model_fitted = modelSARIMAX(dataToFit)
# summary of fit model
print(model_fitted.summary()) 
# line plot of residuals
residuals = pandas.DataFrame(model_fitted.resid) 
residuals.plot()
print("Residuals Plot")
plot.show()
# density plot of residuals
residuals.plot(kind='kde') 
plot.show() 
# summary stats of residuals
print(residuals.describe()) 
```

```python
# predictions
yhat = model_fitted.predict(start=len(dataToFit), end=len(dataToFit) + 24)
plot.plot(yhat)
```

```python
# diagnostics
model_fitted.plot_diagnostics(figsize=(12, 6))
```

```python
# implementing the regression parameters selection based on optimization
# spliting the data into training and test
print(numpy.sum(dataToFit))
print(numpy.mean(dataToFit))
print(dataToFit)
print(len(dataToFit))
train_ts = dataToFit[: 120]
print(len(train_ts))
test_ts = dataToFit[120 :]
print(len(test_ts))
# implementing the optimization model
import pmdarima
from pmdarima import auto_arima
modelo = auto_arima(
            y                 = train_ts,
            start_p           = 0,
            start_q           = 0,
            max_p             = 3,
            max_q             = 3,
            seasonal          = True,
            test              = 'adf',
            m                 = 24, # seasonality
            d                 = None, # apply differentiation
            D                 = None, # apply differentiation on seasonality
            trace             = True,
            error_action      = 'ignore',
            suppress_warnings = True,
            stepwise          = True,
            random_state      = 42, # seed
            n_fits            = 15000
)
print(modelo.summary())
```

```python
# creating a function for fitting SARIMAX
def modelSARIMAX(dataToFit):
     # creating the object model SARIMAX(3,0,1) x (1,0,1)24
    model = SARIMAX(dataToFit, order = (3,0,1), seasonal_order = (1,0,1,24))
    model_fit = model.fit() # fit model
    return model_fit

# calling the function
model_fitted = modelSARIMAX(dataToFit)
# summary of fit model
print(model_fitted.summary()) 
# line plot of residuals
residuals = pandas.DataFrame(model_fitted.resid) 
residuals.plot()
print("Residuals Plot")
plot.show()
# density plot of residuals
residuals.plot(kind='kde') 
plot.show() 
# summary stats of residuals
print(residuals.describe()) 

# predictions
yhat = model_fitted.predict(start=len(dataToFit), end=len(dataToFit) + 24)
plot.plot(yhat)

# diagnostics
model_fitted.plot_diagnostics(figsize=(12, 6))
```
